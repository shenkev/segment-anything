```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.concatenate([i*np.ones(100) for i in range(1, 9)])
```

```python
n = 40
samples = []

for i in range(n):
    s = np.random.choice(a, size=200, replace=False)
    samples.append([sum(s==i) for i in range(1, 9)])
```

```python
plt.hist([s[0] for s in samples], color="blue", alpha=0.5)
```
